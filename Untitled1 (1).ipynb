# Import Libraries

```python
import numpy as np
import os
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import StandardScaler
import pickle
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

# ƒê∆∞·ªùng d·∫´n d·ªØ li·ªáu

```python
DATA_PATH = '/content/drive/MyDrive/Project/XLAS/data_train/'

# ƒê∆∞·ªùng d·∫´n l∆∞u model
MODEL_SAVE_PATH = '/content/drive/MyDrive/Project/XLAS/'
```

 # Load d·ªØ li·ªáu ƒë√£ x·ª≠ l√Ω

```python
print("\n" + "=" * 60)
print("üì• LOADING HOG FEATURES DATA...")
print("=" * 60)

# Load training data
X_train = np.load(DATA_PATH + 'X_train.npy')
y_train = np.load(DATA_PATH + 'y_train.npy')

# Load validation data
X_val = np.load(DATA_PATH + 'X_val.npy')
y_val = np.load(DATA_PATH + 'y_val.npy')

# Load test data
X_test = np. load(DATA_PATH + 'X_test.npy')
y_test = np.load(DATA_PATH + 'y_test.npy')

print(f"\nüìä TRAINING SET:")
print(f"   X_train shape: {X_train.shape}")
print(f"   y_train shape: {y_train.shape}")

print(f"\nüìä VALIDATION SET:")
print(f"   X_val shape: {X_val.shape}")
print(f"   y_val shape: {y_val.shape}")

print(f"\nüìä TEST SET:")
print(f"   X_test shape: {X_test.shape}")
print(f"   y_test shape: {y_test.shape}")

# Hi·ªÉn th·ªã th√¥ng tin classes
unique_classes = np. unique(y_train)
print(f"\nüè∑Ô∏è S·ªë l∆∞·ª£ng classes: {len(unique_classes)}")
print(f"üè∑Ô∏è Danh s√°ch classes: {unique_classes}")

# ƒê·∫øm s·ªë l∆∞·ª£ng m·∫´u m·ªói class
print(f"\nüìà Ph√¢n b·ªë d·ªØ li·ªáu training:")
for cls in unique_classes:
    count = np.sum(y_train == cls)
    print(f"   Class '{cls}': {count} samples")
```

# Chu·∫©n h√≥a d·ªØ li·ªáu (Standardization)

```python
print("\n" + "=" * 60)
print("‚öôÔ∏è STANDARDIZING DATA...")
print("=" * 60)

# Chu·∫©n h√≥a features ƒë·ªÉ SVM ho·∫°t ƒë·ªông t·ªët h∆°n
scaler = StandardScaler()

# Reshape X_train, X_val, X_test from 4D to 2D before scaling
X_train_reshaped = X_train.reshape(X_train.shape[0], -1)
X_val_reshaped = X_val.reshape(X_val.shape[0], -1)
X_test_reshaped = X_test.reshape(X_test.shape[0], -1)

X_train_scaled = scaler.fit_transform(X_train_reshaped)
X_val_scaled = scaler.transform(X_val_reshaped)
X_test_scaled = scaler.transform(X_test_reshaped)

print("‚úÖ Data standardization completed!")
print(f"   Mean of X_train_scaled: {X_train_scaled.mean():.6f}")
print(f"   Std of X_train_scaled: {X_train_scaled.std():.6f}")
```

#  K·∫øt h·ª£p train v√† validation ƒë·ªÉ training

```python
X_train_full = np.vstack([X_train_scaled, X_val_scaled])
y_train_full = np.concatenate([y_train, y_val])

print(f"\nüìä COMBINED TRAINING SET (Train + Val):")
print(f"   X_train_full shape: {X_train_full. shape}")
print(f"   y_train_full shape: {y_train_full.shape}")
```

# Training SVM Model

```python

print("\n" + "=" * 60)
print("üöÄ TRAINING SVM MODEL...")
print("=" * 60)

# Kh·ªüi t·∫°o SVM classifier
# C√°c hyperparameters ph·ªï bi·∫øn cho nh·∫≠n di·ªán k√Ω t·ª±:
# - kernel='rbf': Ph√π h·ª£p cho d·ªØ li·ªáu phi tuy·∫øn
# - C=10: Regularization (tƒÉng C = gi·∫£m regularization)
# - gamma='scale': T·ª± ƒë·ªông t√≠nh gamma d·ª±a tr√™n features

svm_model = SVC(
    kernel='rbf',           # Radial Basis Function kernel
    C=10,                   # Regularization parameter
    gamma='scale',          # Kernel coefficient
    decision_function_shape='ovr',  # One-vs-Rest cho multi-class
    random_state=42,
    verbose=True
)

# B·∫Øt ƒë·∫ßu training
start_time = datetime. now()
print(f"\n‚è∞ Training started at: {start_time. strftime('%H:%M:%S')}")

# Convert y_train_full from one-hot encoding to label encoding
y_train_full_labels = np.argmax(y_train_full, axis=1)

# Reshape X_train_full from 4D to 2D before training
X_train_full_reshaped = X_train_full.reshape(X_train_full.shape[0], -1)

svm_model.fit(X_train_full_reshaped, y_train_full_labels)

end_time = datetime. now()
training_time = end_time - start_time
print(f"\n‚úÖ Training completed!")
print(f"‚è±Ô∏è Training time: {training_time}")
```

#  ƒê√°nh gi√° Model tr√™n Validation Set

```python
print("\n" + "=" * 60)
print("üìä EVALUATING ON VALIDATION SET...")
print("=" * 60)

y_val_pred = svm_model.predict(X_val_scaled)

# Convert y_val from one-hot encoding to label encoding
y_val_labels = np.argmax(y_val, axis=1)

val_accuracy = accuracy_score(y_val_labels, y_val_pred)
print(f"\nüéØ Validation Accuracy: {val_accuracy * 100:.2f}%")
```

# ƒê√°nh gi√° Model tr√™n Test Set

```python
print("\n" + "=" * 60)
print("üìä EVALUATING ON TEST SET...")
print("=" * 60)

# D·ª± ƒëo√°n tr√™n t·∫≠p test
y_test_pred = svm_model.predict(X_test_scaled)

# Convert y_test from one-hot encoding to label encoding
y_test_labels = np.argmax(y_test, axis=1)

# T√≠nh accuracy
test_accuracy = accuracy_score(y_test_labels, y_test_pred)
print(f"\nüéØ Test Accuracy: {test_accuracy * 100:.2f}%")

# Classification report chi ti·∫øt
print("\n" + "-" * 40)
print("üìã CLASSIFICATION REPORT:")
print("-" * 40)
print(classification_report(y_test_labels, y_test_pred))
```

# Visualize Confusion Matrix

```python
print("\n" + "=" * 60)
print("üìä GENERATING CONFUSION MATRIX...")
print("=" * 60)

# T·∫°o confusion matrix, s·ª≠ d·ª•ng y_test_labels ƒë√£ ƒë∆∞·ª£c chuy·ªÉn ƒë·ªïi
cm = confusion_matrix(y_test_labels, y_test_pred)

# L·∫•y danh s√°ch c√°c l·ªõp duy nh·∫•t t·ª´ y_test_labels ƒë·ªÉ d√πng cho nh√£n tr·ª•c
plot_class_labels = np.unique(y_test_labels)

# Plot
plt.figure(figsize=(14, 12))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=plot_class_labels,
            yticklabels=plot_class_labels,
            annot_kws={'size': 8})
plt.title('Confusion Matrix - SVM + HOG License Plate Recognition\n' +
          f'Test Accuracy: {test_accuracy * 100:.2f}%', fontsize=14)
plt.xlabel('Predicted Label', fontsize=12)
plt.ylabel('True Label', fontsize=12)
plt. xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()

# L∆∞u h√¨nh
cm_path = MODEL_SAVE_PATH + 'confusion_matrix_svm_hog.png'
plt.savefig(cm_path, dpi=150, bbox_inches='tight')
plt.show()

print(f"‚úÖ Confusion matrix saved to: {cm_path}")
```

# L∆∞u Model v√† Scaler

```python
print("\n" + "=" * 60)
print("üíæ SAVING MODEL AND SCALER...")
print("=" * 60)

# L∆∞u SVM model
model_path = MODEL_SAVE_PATH + 'svm_hog_license_plate_model.pkl'
with open(model_path, 'wb') as f:
    pickle.dump(svm_model, f)
print(f"‚úÖ SVM Model saved to: {model_path}")

# L∆∞u Scaler (quan tr·ªçng ƒë·ªÉ s·ª≠ d·ª•ng khi predict)
scaler_path = MODEL_SAVE_PATH + 'scaler. pkl'
with open(scaler_path, 'wb') as f:
    pickle.dump(scaler, f)
print(f"‚úÖ Scaler saved to: {scaler_path}")

# L∆∞u danh s√°ch classes
classes_path = MODEL_SAVE_PATH + 'classes.npy'
np. save(classes_path, unique_classes)
print(f"‚úÖ Classes saved to: {classes_path}")
```

# Test Model ƒë√£ l∆∞u

```python
print("\n" + "=" * 60)
print("üß™ TESTING SAVED MODEL...")
print("=" * 60)

# Load model
with open(model_path, 'rb') as f:
    loaded_model = pickle.load(f)

# Load scaler
with open(scaler_path, 'rb') as f:
    loaded_scaler = pickle.load(f)

# Test v·ªõi m·ªôt s·ªë samples ng·∫´u nhi√™n
print("\nüîç Random Sample Predictions:")
print("-" * 50)

np.random.seed(42)
sample_indices = np.random.choice(len(X_test), 10, replace=False)

correct = 0
for i, idx in enumerate(sample_indices):
    sample = X_test[idx]. reshape(1, -1)
    sample_scaled = loaded_scaler.transform(sample)
    prediction = loaded_model. predict(sample_scaled)

    # Convert actual from one-hot encoding to a single label
    actual_label = np.argmax(y_test[idx])

    is_correct = prediction[0] == actual_label
    if is_correct:
        correct += 1
    status = "‚úÖ" if is_correct else "‚ùå"

    print(f"   Sample {i+1}: Predicted = '{prediction[0]}' | Actual = '{actual_label}' {status}")

print(f"\nüìä Sample Accuracy: {correct}/{len(sample_indices)} = {correct/len(sample_indices)*100:.1f}%")
```
