# Import Libraries

```python
import numpy as np
import cv2
from skimage.feature import hog
from sklearn.svm import LinearSVC, SVC # Added SVC here
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import pickle
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

# ƒê∆∞·ªùng d·∫´n d·ªØ li·ªáu

```python
DATA_PATH = '/content/drive/MyDrive/Project/XLAS/data_train/'

# ƒê∆∞·ªùng d·∫´n l∆∞u model
MODEL_SAVE_PATH = '/content/drive/MyDrive/Project/XLAS/'
```

 # Load d·ªØ li·ªáu ƒë√£ x·ª≠ l√Ω

```python
print("=" * 60)
print("üì• LOADING DATA...")
print("=" * 60)

# Load image data
X_train = np.load(DATA_PATH + 'X_train.npy')
y_train = np.load(DATA_PATH + 'y_train.npy')
X_val = np.load(DATA_PATH + 'X_val.npy')
y_val = np.load(DATA_PATH + 'y_val.npy')
X_test = np.load(DATA_PATH + 'X_test.npy')
y_test = np.load(DATA_PATH + 'y_test.npy')

# ‚≠ê Load class weights
class_weights_array = np.load(DATA_PATH + 'class_weights.npy')

print(f"\nüìä Data shapes:")
print(f"   X_train: {X_train.shape}")
print(f"   X_val: {X_val. shape}")
print(f"   X_test: {X_test.shape}")

print(f"\n‚öñÔ∏è Class weights:")
print(f"   Shape: {class_weights_array. shape}")
print(f"   Values: {class_weights_array}")

# Convert one-hot to labels
y_train_labels = np.argmax(y_train, axis=1)
y_val_labels = np.argmax(y_val, axis=1)
y_test_labels = np.argmax(y_test, axis=1)

n_classes = y_train. shape[1]
print(f"\nüè∑Ô∏è Number of classes: {n_classes}")
```

# T·∫°o Class Weight Dictionary

```python
print("\n" + "=" * 60)
print("‚öñÔ∏è CREATING CLASS WEIGHT DICTIONARY...")
print("=" * 60)

# X·ª≠ l√Ω class_weights_array t√πy theo format
if len(class_weights_array.shape) == 1:
    # Format: [w0, w1, w2, ... ] - m·ªói ph·∫ßn t·ª≠ l√† weight c·ªßa class t∆∞∆°ng ·ª©ng
    if len(class_weights_array) == n_classes:
        class_weight_dict = {i: class_weights_array[i] for i in range(n_classes)}
    else:
        # C√≥ th·ªÉ l√† format kh√°c, t√≠nh to√°n l·∫°i
        print("‚ö†Ô∏è Class weights length kh√¥ng kh·ªõp v·ªõi s·ªë classes, t√≠nh to√°n l·∫°i...")
        from sklearn.utils.class_weight import compute_class_weight
        class_weight_array_computed = compute_class_weight(
            class_weight='balanced',
            classes=np.unique(y_train_labels),
            y=y_train_labels
        )
        class_weight_dict = {i: class_weight_array_computed[i] for i in range(len(class_weight_array_computed))}
else:
    # Format 2D: [[class_id, weight], ...]
    class_weight_dict = {int(row[0]): row[1] for row in class_weights_array}

print(f"‚úÖ Class weight dictionary created!")
print(f"   Number of classes: {len(class_weight_dict)}")
print(f"   Sample: Class 0 = {class_weight_dict.get(0, 'N/A'):.4f}, Class 1 = {class_weight_dict.get(1, 'N/A'):.4f}")
```

# Tr√≠ch xu·∫•t HOG Features

```python
print("\n" + "=" * 60)
print("üîÑ EXTRACTING HOG FEATURES...")
print("=" * 60)

def extract_hog_features(images, name=""):
    """
    Tr√≠ch xu·∫•t HOG features t·ª´ t·∫≠p ·∫£nh
    """
    hog_features = []
    total = len(images)

    print(f"\nüì∑ Processing {name}: {total} images")

    for i, img in enumerate(images):
        # Hi·ªÉn th·ªã ti·∫øn tr√¨nh
        if (i + 1) % 5000 == 0 or i == 0 or i == total - 1:
            print(f"   [{i+1}/{total}] - {((i+1)/total*100):.1f}%")

        # Chuy·ªÉn sang grayscale
        if len(img. shape) == 3:
            gray = cv2.cvtColor(img. astype(np.uint8), cv2.COLOR_RGB2GRAY)
        else:
            gray = img.astype(np.uint8)

        # Tr√≠ch xu·∫•t HOG features
        features = hog(
            gray,
            orientations=9,           # S·ªë h∆∞·ªõng gradient
            pixels_per_cell=(8, 8),   # K√≠ch th∆∞·ªõc cell
            cells_per_block=(2, 2),   # S·ªë cells m·ªói block
            block_norm='L2-Hys',      # Chu·∫©n h√≥a block
            visualize=False,
            feature_vector=True
        )

        hog_features.append(features)

    return np.array(hog_features)

# Tr√≠ch xu·∫•t HOG
start_time = datetime.now()

X_train_hog = extract_hog_features(X_train, "X_train")
X_val_hog = extract_hog_features(X_val, "X_val")
X_test_hog = extract_hog_features(X_test, "X_test")

print(f"\n‚úÖ HOG extraction completed!")
print(f"‚è±Ô∏è Time: {datetime.now() - start_time}")
print(f"\nüìä HOG feature shapes:")
print(f"   X_train_hog: {X_train_hog.shape}")
print(f"   X_val_hog: {X_val_hog.shape}")
print(f"   X_test_hog: {X_test_hog.shape}")
```

#  Chu·∫©n h√≥a d·ªØ li·ªáu

```python
print("\n" + "=" * 60)
print("‚öôÔ∏è STANDARDIZING HOG FEATURES...")
print("=" * 60)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_hog)
X_val_scaled = scaler.transform(X_val_hog)
X_test_scaled = scaler.transform(X_test_hog)

print("‚úÖ Standardization completed!")
print(f"   Mean: {X_train_scaled.mean():.6f}")
print(f"   Std: {X_train_scaled.std():.6f}")
```

# K·∫øt h·ª£p Train + Val

```python
X_train_full = np.vstack([X_train_scaled, X_val_scaled])
y_train_full = np.concatenate([y_train_labels, y_val_labels])

print(f"\nüìä Combined training set:")
print(f"   X_train_full: {X_train_full.shape}")
print(f"   y_train_full: {y_train_full.shape}")
```

# Training SVM Model

```python
print("\n" + "=" * 60)
print("üöÄ TRAINING LinearSVC WITH HOG + CLASS WEIGHTS...")
print("=" * 60)

# ‚≠ê S·ª≠ d·ª•ng LinearSVC v·ªõi class_weight
svm_model = LinearSVC(
    C=1.0,
    class_weight=class_weight_dict,  # ‚≠ê S·ª¨ D·ª§NG CLASS WEIGHTS
    max_iter=10000,
    dual=False,
    random_state=42,
    verbose=1
)

start_time = datetime.now()
print(f"\n‚è∞ Training started at: {start_time. strftime('%H:%M:%S')}")

svm_model.fit(X_train_full, y_train_full)

end_time = datetime.now()
training_time = end_time - start_time
print(f"\n‚úÖ Training completed!")
print(f"‚è±Ô∏è Training time: {training_time}")
```

#  ƒê√°nh gi√° Model tr√™n Validation Set

```python
print("\n" + "=" * 60)
print("üìä EVALUATING ON VALIDATION SET...")
print("=" * 60)

y_val_pred = svm_model.predict(X_val_scaled)
val_accuracy = accuracy_score(y_val_labels, y_val_pred)
print(f"\nüéØ Validation Accuracy: {val_accuracy * 100:.2f}%")
```

# ƒê√°nh gi√° Model tr√™n Test Set

```python
print("\n" + "=" * 60)
print("üìä EVALUATING ON TEST SET...")
print("=" * 60)

y_test_pred = svm_model.predict(X_test_scaled)
test_accuracy = accuracy_score(y_test_labels, y_test_pred)
print(f"\nüéØ Test Accuracy: {test_accuracy * 100:.2f}%")

# Gap
gap = (val_accuracy - test_accuracy) * 100
print(f"\nüìä Gap (Val - Test): {gap:.2f}%")

print("\n" + "-" * 40)
print("üìã CLASSIFICATION REPORT:")
print("-" * 40)
print(classification_report(y_test_labels, y_test_pred))
```

# Visualize Confusion Matrix

```python
print("\n" + "=" * 60)
print("üìä CONFUSION MATRIX...")
print("=" * 60)

cm = confusion_matrix(y_test_labels, y_test_pred)

plt.figure(figsize=(14, 12))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', annot_kws={'size': 7})
plt.title(f'Confusion Matrix - HOG + SVM + Class Weights\n'
          f'Validation: {val_accuracy*100:.2f}% | Test: {test_accuracy*100:.2f}% | Gap: {gap:.2f}%',
          fontsize=14)
plt.xlabel('Predicted Label', fontsize=12)
plt.ylabel('True Label', fontsize=12)
plt.tight_layout()
plt.savefig(MODEL_SAVE_PATH + 'confusion_matrix_hog_cw.png', dpi=150)
plt.show()

print(f"‚úÖ Saved to: {MODEL_SAVE_PATH}confusion_matrix_hog_cw.png")

```

# L∆∞u Model v√† Scaler

```python
print("\n" + "=" * 60)
print("üíæ SAVING MODEL AND DATA...")
print("=" * 60)

# L∆∞u SVM model
model_path = MODEL_SAVE_PATH + 'svm_hog_classweight_model.pkl'
with open(model_path, 'wb') as f:
    pickle.dump(svm_model, f)
print(f"‚úÖ Model saved: {model_path}")

# L∆∞u Scaler
scaler_path = MODEL_SAVE_PATH + 'hog_scaler.pkl'
with open(scaler_path, 'wb') as f:
    pickle.dump(scaler, f)
print(f"‚úÖ Scaler saved: {scaler_path}")

# L∆∞u HOG features (ƒë·ªÉ kh√¥ng c·∫ßn extract l·∫°i)
np.save(MODEL_SAVE_PATH + 'X_train_hog.npy', X_train_hog)
np.save(MODEL_SAVE_PATH + 'X_val_hog.npy', X_val_hog)
np.save(MODEL_SAVE_PATH + 'X_test_hog.npy', X_test_hog)
print(f"‚úÖ HOG features saved!")
```

# Test Model ƒë√£ l∆∞u

```python
print("\n" + "=" * 60)
print("üß™ TESTING SAVED MODEL...")
print("=" * 60)

# Load model
with open(model_path, 'rb') as f:
    loaded_model = pickle.load(f)

# Load scaler
with open(scaler_path, 'rb') as f:
    loaded_scaler = pickle.load(f)

# Test v·ªõi m·ªôt s·ªë samples ng·∫´u nhi√™n
print("\nüîç Random Sample Predictions:")
print("-" * 50)

np.random.seed(42)
sample_indices = np.random.choice(len(X_test), 10, replace=False)

correct = 0
for i, idx in enumerate(sample_indices):
    sample = X_test[idx]. reshape(1, -1)
    sample_scaled = loaded_scaler.transform(sample)
    prediction = loaded_model. predict(sample_scaled)

    # Convert actual from one-hot encoding to a single label
    actual_label = np.argmax(y_test[idx])

    is_correct = prediction[0] == actual_label
    if is_correct:
        correct += 1
    status = "‚úÖ" if is_correct else "‚ùå"

    print(f"   Sample {i+1}: Predicted = '{prediction[0]}' | Actual = '{actual_label}' {status}")

print(f"\nüìä Sample Accuracy: {correct}/{len(sample_indices)} = {correct/len(sample_indices)*100:.1f}%")
```
